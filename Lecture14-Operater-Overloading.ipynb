```python
class Phone:
    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.price = price
    
    def __add__(self, other):
        return self.price + other.price

    def __sub__(self, other):
        return self.price - other.price

    def __mul__(self, other):
        return self.price * other.price

    def __pow__(self, other):
        return self.price ** other.price

    def __truediv__(self, other):
        return self.price / other.price
```

```python
phone1 = Phone('IPhone', 'XR', 40000)
```

```python
phone1.__dict__
```

```python
phone2 = Phone('Samsung', 'A15', 20000)
```

```python
phone1 + phone2    #__add__
```

```python
phone1 - phone2
```

```python
phone1 ** phone2
```

```python
phone1 / phone2
```
